```python
data.rename(columns=lambda x:x.strip(),inplace=True)
```

```python
data
```

```python
data['Arrival_Date']=pd.to_datetime(data['Arrival_Date'],errors='coerce',dayfirst=True)
```

```python
data=data.sort_values(by=['Market','Commodity','Arrival_Date'])
```

```python
data
```

```python
data['ModalPrice_PctChange'] = data.groupby(['Market', 'Commodity'])['Modal Price'].pct_change() * 100
```

```python
data
```

```python
data.tail()
```

```python
data = data.dropna(subset=['ModalPrice_PctChange'])
```

```python
data.info()
```

```python
data.describe()
```

```python
data = data[data['Modal Price'] > 0]
```

```python
data.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data['scaled_ModalPrice_PctChange']=scaler.fit_transform(data[['ModalPrice_PctChange']])
```

```python
data
```

```python
from sklearn.ensemble import IsolationForest

X = data[['scaled_ModalPrice_PctChange']]
model = IsolationForest(contamination=0.05, random_state=42)
data['Anomaly'] = model.fit_predict(X)
data['Anomaly'] = data['Anomaly'].map({1: "Normal", -1: "Anomaly"})

```
